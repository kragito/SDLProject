```python
#First of all, I import all of the packages required for this assignment
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
df_MR = pd.read_csv('age-standardised-mortality-rate-for-ischaemic-heart-disease.csv', index_col=0)
df_CMR = pd.read_csv('age-standardised-mortality-rate-for-cancer.csv', index_col=0)
df_SMR = pd.read_csv('age-standardised-mortality-rate-for-stroke.csv', index_col=0)
MRY = df_MR.index
CIHD = df_CMR['cancer']
SIHD = df_SMR['stroke']
IHD = df_MR['ihd']
plt.scatter(MRY,IHD) #plots scatter graph
plt.scatter(MRY,CIHD)
plt.scatter(MRY,SIHD)
plt.legend(['Heart Disease', 'Cancer', 'Stroke']);
plt.ylabel('Deaths per 100,000')
plt.xlabel('Year')
plt.show()

x1 = IHD
x2 = CIHD
x3 = SIHD
# Assign colors for each illness and the names
colors = ['blue', 'green', 'red']
names = ['Heart Disease.', 'Cancer', 'Stroke']
         
# Make the histogram using a list of lists
plt.hist([x1, x2, x3], bins = 10,
         color = colors, label=names)

# Plot formatting
plt.legend()
plt.xlabel('Mortality Rate')
plt.title('Side-by-Side Histogram with All Illnesses')
```
